###Контрольные вопросы:
*1.Типы ячеек в Jupyter Notebook?* - Ячейки кода (Code Cells): Ячейки, в которых пишется исполняемый код на Python или другом языке программирования. Результаты выполнения отображаются сразу под ячейкой.Текстовые ячейки (Markdown Cells): Ячейки для написания текстового контента с использованием Markdown. Они позволяют форматировать текст, добавлять заголовки, списки и изображения.<br>
*2.Как с помощью Markdown сформировать иерархический список?* - Для создания иерархического списка в Markdown используйте символы *, - или + для ненумерованных списков и цифры для нумерованных списков. Для вложения элементов просто добавьте отступы.<br>
3.Как с помощью Markdown оформить цитату?* - Чтобы оформить цитату в Markdown, поставьте символ "&gt"; перед текстом. Это выделяет текст как цитату. <br>
*4.Как в текстовую ячейку вставить отформатированный блок кода?* - Для вставки кода в текстовую ячейку используйте тройные обратные кавычки (```) или отступ в 4 пробела.<br>
*5.Сколько уровней заголовка существует в Markdown?* - В Markdown существует 6 уровней заголовков, которые обозначаются с помощью символа #.
# Заголовок 1
## Заголовок 2
### Заголовок 3
#### Заголовок 4
##### Заголовок 5
###### Заголовок 6

*7.Типы данных, рассказать подробно про один из типов данных (на выбор преподавателя)?* - int, float, str, list, tuple, set, dict.<br>
*8.Что такое Jupyter Notebook, для чего используется?* - Jupyter Notebook — это веб-приложение для создания и обмена документами, содержащими код, уравнения, визуализации и объяснительный текст. Он широко используется в научных исследованиях, обучении, разработке алгоритмов и анализе данных, позволяя интерактивно выполнять код и визуализировать результаты на одном и том же интерфейсе.<br>
*9.Почему python популярен для анализа данных?* - Простота и читаемость синтаксиса, что облегчает изучение и использование языка. Огромное количество библиотек для научных вычислений и анализа данных, таких как NumPy, Pandas, Matplotlib и Scikit-learn. Широкое сообщество и поддержка, что облегчает поиск решений и ресурсов.<br>
*10.Обязательно ли использовать Jupyter Notebook для анализа данных?* - Нет, Jupyter Notebook не является обязательным инструментом для анализа данных. Анализ можно проводить с помощью различных IDE (например, PyCharm, VSCode) или командной строки. Однако Jupyter Notebook предоставляет множество возможностей для интерактивной работы и визуализации, что делает его популярным выбором среди аналитиков и исследователей.
