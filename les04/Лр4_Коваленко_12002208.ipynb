6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
mol=titanic_df[titanic_df["Age"] <30]["Survived"]
summ=mol.count()
sumsur=mol.value_counts().iloc[1]
print("младше 30= ",sumsur/summ*100)
oldp=titanic_df[titanic_df["Age"] >60]["Survived"]
summ=oldp.count()
sumsur=oldp.value_counts().iloc[1]
print("старше 60= ",round(sumsur/summ*100,3))
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
mol=titanic_df[titanic_df["Sex"] == 'male']["Survived"]
summ=mol.count()
sumsur=mol.value_counts().iloc[1]
print("выжившие мужчины= ",round(sumsur/summ*100,2))
oldp=titanic_df[titanic_df["Sex"] == 'female']["Survived"]
summ=oldp.count()
sumsur=oldp.value_counts().iloc[1]
print("выжившие женщины= ",round(sumsur/summ*100,2))
```



9. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

    В среднем мужчины 1-го класса старше 40 лет.
    В среднем женщины 1-го класса старше 40 лет.
    Мужчины всех классов в среднем старше женщин того же класса.
    В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса.



```python
z1= titanic_df[(titanic_df["Sex"] == 'male') & (titanic_df["Pclass"] == 1)]["Age"].mean()>40
print('1 zadanie:',z1)
z2= titanic_df[(titanic_df["Sex"] == 'female') & (titanic_df["Pclass"] == 1)]["Age"].mean()>40
print('2 zadanie:',z2)
mens= titanic_df[(titanic_df["Sex"] == 'male')].groupby('Pclass')["Age"].mean()
shes = titanic_df[(titanic_df["Sex"] == 'female')].groupby('Pclass')["Age"].mean()
print('3 zadanie:',(mens>shes).all())
shes = titanic_df.groupby('Pclass')["Age"].mean()
print('3 zadanie:',shes.loc[1]>shes.loc[2]>shes.loc[3])
```
